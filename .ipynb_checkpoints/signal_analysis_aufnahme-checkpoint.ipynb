```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from pydub import AudioSegment
from scipy.fft import fft, fftfreq, rfft, rfftfreq
from scipy.io import wavfile
from mplcursors import cursor
```

```python
audio = AudioSegment.from_file("aufnahme.m4a", format="m4a")
signal = np.array(audio.get_array_of_samples())
sample_rate = audio.frame_rate
```

```python
if len(signal.shape) == 2:
    signal = signal.mean(axis=1)
```

```python
n = len(signal)
duration = n / sample_rate
t = np.linspace(0.0, duration, n, endpoint=False)
```

```python
yf = fft(signal)
xf = fftfreq(n, 1 / sample_rate)
```

```python
positive_freqs = xf[:n // 2]
magnitude = 2.0 / n * np.abs(yf[:n // 2])
```

```python
plt.figure(figsize=(10, 6))
a=plt.plot(positive_freqs, magnitude)
plt.title('Frequenzspektrum der WAV-Datei')
plt.xlabel('Frequenz (Hz)')
plt.ylabel('Amplitude')
plt.grid()
cursor(a)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, signal)
plt.title('Zeitbereich des Signals')
plt.xlabel('Zeit (Sekunden)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```
