```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from pydub import AudioSegment
from scipy.fft import rfft, rfftfreq
from mplcursors import cursor
```

```python
audio = AudioSegment.from_file('./audiofiles/aufnahme.m4a', format="m4a")
signal = np.array(audio.get_array_of_samples())
sample_rate = audio.frame_rate
```

```python
if len(signal.shape) == 2:
    signal = signal.mean(axis=1)
```

```python
n = len(signal)
duration = n / sample_rate
t = np.linspace(0.0, duration, n, endpoint=False)
```

```python
yf = rfft(signal) / n
xf = rfftfreq(n, 1 / sample_rate)
```

```python
#positive_freqs = xf[:n // 2]
magnitude = np.abs(yf)
```

```python
plt.figure(figsize=(10, 6))
a=plt.plot(xf, magnitude)
plt.title('Frequenzspektrum der M4A-Datei')
plt.xlabel('Frequenz (Hz)')
plt.ylabel('Amplitude')
plt.grid()
cursor(a)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, signal)
plt.title('Zeitbereich des Signals')
plt.xlabel('Zeit (Sekunden)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```
